```python
# Series of counts for a brand/product
top_5_count = data['brand_name'].value_counts()[0:5]

top_5_price = []

for brand in top_5_count.index:
    top_5_price.append( data[data['brand_name'] == brand]['price_usd'].values )
```

```python
# Create figure and axes
fig, ax = plt.subplots(1,2, figsize = (10,5))

################## AX 0 ###################
# store in variable to create a bar_label object in next line
bc = ax[0].barh(top_5_count.index, top_5_count.values)
# Add count total to bars
ax[0].bar_label(bc, label_type = 'center')
# Add title, xlabel
ax[0].set(title = 'Product Counts', xlabel = 'Count')
# Change fontsize for y axis
ax[0].tick_params(axis='y', which='major', labelsize=10)
# Rename yticks: Shorten "SEPHORA COLLECTION" TO "SEPHORA" etc
ax[0].set_yticks(ax[0].get_yticks(), labels = ['SEPHORA','CLINIQUE', 'Dior', 'tarte', 'NEST'])



################## AX 1 ###################
''' Boxplot accepts nested arrays of different sizes to create different categories of the boxplot'''
ax[1].boxplot(x = top_5_price, vert = False); ax[1].set_yticks([])
ax[1].set(title = 'Price Distribution', xlabel = 'USD ($)');
```
